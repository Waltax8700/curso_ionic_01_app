#*Cambiar el espacio de color*

OpenCV proporciona la función cvtColor () para convertir la imagen de un espacio de color a otro. Toma la imagen como entrada y también el código de conversión del espacio de color. Los siguientes son algunos códigos de conversión:

COLOR_BGR2GRAY 

COLOR_BGR2HSV 

COLOR_HSV2BGR 

COLOR_BGR2YUV 

COLOR_GRAY2BGR


El siguiente es un código de ejemplo para convertir una imagen BGR en una imagen en escala de grises. 

Para mostrar una imagen en escala de grises, configure el mapeo de colores utilizando los parámetros cmap = 'gray', vmin = 0, vmax = 255.

Podemos cambiar el segundo parámetro de la función cvtColor () por cualquiera de los valores de parámetro mencionados anteriormente:

```python
# Cambio de formato de la imagen de BGR a Gray Scale
img_cvt1=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Mostrar la imagen
plt.imshow(img_cvt1,cmap = 'gray', vmin = 0, vmax = 255)
plt.show()
print("Imagen modificada a escala de grises")
#cv2.imwrite("img_gray.jpg", img_cvt1)
```

El siguiente es un código de ejemplo para convertir una imagen BGR en una imagen RGB:

```python
# Cambio de formato de la imagen de BGR a RGB
img_cvt2=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Mostrar la imagen
plt.imshow(img_cvt2)
plt.show()
print("Imagen modificada a RGB")
```

#*Espacio de color HSV* 

Representación tridimensional del color basado en los componentes de tinte, matiz o tonalidad (hue, en inglés), saturación (saturation) y brillo o valor (value). Fue definido en 1978 por Alvy Ray Smith.

A diferencia del modelo RGB ampliamente usado en los monitores, televisores, etc., si bien las coordenadas de aquel son euclideanas; el color HSV sigue una representación más parecida a las coordenadas cilíndricas. Además es una representación más cercana a la forma en que los humanos perciben los colores y sus propiedades, pues se agrupan las tonalidades de color, lo cual es distinto al caso RGB donde los colores no están necesariamente tan agrupados. Se diferencia con HSL en el uso del componente de saturación.

Utilizando el espacio de color HSV. El espacio de color HSV se utiliza principalmente para el seguimiento de objetos.

```python
# Cambio de formato de la imagen de BGR a RGB
img_cvt3=cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
# Mostrar la imagen
plt.imshow(img_cvt3)
plt.show()
print("Imagen modificada a HSV")
```

#*Escalada*

Para cambiar el tamaño de la imagen, OpenCV tiene una función resize (), que toma la imagen, las dimensiones y el algoritmo de interpolación como entrada. Se pueden utilizar varios algoritmos de interpolación para interpolar los nuevos valores de píxeles. Los siguientes son los algoritmos de interpolación, que se pueden utilizar para cambiar el tamaño de la imagen:


cv2.INTER_AREA: Se prefiere este algoritmo para reducir la imagen cv2.

INTER_CUBIC: Se prefiere este algoritmo para hacer zoom (lento) cv2.

INTER_LINEAR: Este algoritmo se prefiere para hacer zoom cv2.INTER_LINEAR: Este es el algoritmo predeterminado


El siguiente es un ejemplo de cómo cambiar el tamaño de una imagen:

```python
# Cambio de tamaño de la imagen
r,c = img_cvt2.shape[:2]
new_img = cv2.resize(img_cvt2, (2*c,2*r), interpolation = cv2.INTER_CUBIC)

# Mostrar la imagen original
plt.imshow(img_cvt2)
plt.show()
print("Imagen original en tamaño")
print (r,c)

# Mostrar la imagen modificad
plt.imshow(new_img)
plt.show()
print("Imagen modificada en tamaño")
print (2*c,2*r)
```

#*Recortando la imagen*

Recortar una imagen en OpenCV es muy fácil. Se puede hacer cortando la matriz de imágenes. Cortar una matriz es simplemente tomar los valores de la matriz dentro de valores de índice particulares. Considere el siguiente ejemplo:

```python
# Recorte de la imagen
img_crop = img_cvt2[0:200, 150:350]

# Mostrar la imagen
plt.imshow(img_crop)
plt.show()
print("Imagen recortada")
```

#*Traslación*

Para la transformación geométrica, OpenCV proporciona la función wrapAffine (), que toma la imagen, la matriz de transformación y la dimensión de la imagen como entrada. La matriz de transformación para la traslación es la siguiente:

M = [1 0 tx;  0 1 ty]

Aquí, (tx, ty) nos dice la cantidad de desplazamiento en la imagen. Por ejemplo, el siguiente código cambia la imagen en (100,100):

```python
# Traslación de la imagen
r,c = img.shape[:2]
M = np.float32([[1,0,100],[0,1,100]])
new_img = cv2.warpAffine(img_cvt2,M,(c,r))


# Mostrar la imagen
plt.imshow(new_img)
plt.show()
print("Imagen recortada y trasladada")
```

#*Rotación*

La rotación también se puede hacer usando wrapAffine () - solo cambia la matriz de transformación. La matriz de transformación para la rotación es la siguiente:

M = [cos(theta)   -sen(theta) ; sen(theta)   cos(theta)]

Aquí, theta es el ángulo con el que desea rotar la imagen. El siguiente código es un ejemplo de rotación de la imagen 90 grados, En este código, usamos la función getRotationMatrix2D () para generar la matriz de transformación. Toma el centro de rotación, el ángulo de rotación y el factor de escala como entrada. Este es el resultado del código anterior:

```python
# Rotación de la imagen
r,c = img_cvt2.shape[:2]
M = cv2.getRotationMatrix2D((c/2,r/2),90,1)
new_img = cv2.warpAffine(img_cvt2,M,(c,r))

# Mostrar la imagen
plt.imshow(new_img)
plt.show()
print("Imagen recortada y rotada 90°")
```
